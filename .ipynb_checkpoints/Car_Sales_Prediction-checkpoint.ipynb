# Data Analysis/Visualization

### import all the required libraries

```python
# import all the required libraries
import numpy as np
import pandas as pd
import matplotlib as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.simplefilter('ignore')
```

### read the dataset

```python
df = pd.read_csv('Car_Purchasing_Data.csv', encoding='latin')
df.head()
```

### Descriptive Statistics

```python
df.describe(include='all')
```

### Missing value imputation

```python
df.isnull().sum()
```

### Graphical representation

### Univariant Analysis

```python
# Categorical Data
import matplotlib.pyplot as plt
plt.subplot(221)
df['Gender'].value_counts().plot(kind='bar', title='Gender of students', figsize=(16,8))

plt.subplot(222)
df['Gender'].value_counts().plot(kind="pie", autopct='%.2f')

#plt.savefig('Visualization/EDA_process/barplot_piechart_gender.jpg')
```

### Histogram 

```python
# Numerical Data
plt.hist(df['Age'],bins=10)
#plt.savefig('Visualization/EDA_process/histogram_age.jpg')
```

### Distoplot

```python
sns.distplot(df['Annual Salary'])
df['Annual Salary'].skew()
#plt.savefig('Visualization/EDA_process/distoplot_annual_salary.jpg')
```

### Boxplot

```python
sns.boxplot(df['Car Purchase Amount'])
#plt.savefig('Visualization/EDA_process/boxplot_car_purchase_amount.jpg')
```

### Bivariante and Multivariante Analysis

```python
# Scatterplot (Numerical to Numerical)
sns.scatterplot(df['Annual Salary'],df['Car Purchase Amount'])
#plt.savefig('Visualization/EDA_process/scatterplot_AnnualSalary_CarPurchase.jpg')
```

### 1. Scatterplot

```python
sns.scatterplot(df['Annual Salary'],df['Car Purchase Amount'],hue=df['Gender'])
#plt.savefig('Visualization/EDA_process/scatterplot_AnnualSalary_CarPurchase_Gender.jpg')
```

### 2. Barplot (numerical - categorical)

```python
sns.barplot(df['Gender'],df['Car Purchase Amount'])
#plt.savefig('Visualization/EDA_process/boxplot_car_purchase_amount_Gender.jpg')
```

### 3. Boxplot (numerical - categorical)

```python
sns.boxplot(df['Gender'],df['Car Purchase Amount'])
#plt.savefig('Visualization/EDA_process/barplot_CarPurchase_Gender.jpg')
```

### 5. Heatmap - target variable


```python
corr = df.corr()
sns.set(rc = {'figure.figsize':(16,8)})
sns.heatmap(corr, annot=True, square=True)
#plt.savefig('Visualization/EDA_process/heatmap.jpg')
```

### 7. Pairplot

```python
df.head()
```

```python
sns.pairplot(df, hue='Gender')
#plt.savefig('Visualization/EDA_process/pairplot.jpg')
```

### Lineplot (Numerical - Numercical)

```python
new = df.groupby('Age').sum().reset_index()
new
```

```python
sns.lineplot(new['Age'], new['Car Purchase Amount'])
#plt.savefig('Visualization/EDA_process/lineplot_age_carPurchase.jpg')
```
